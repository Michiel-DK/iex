```python
# Import iex package

from stt.iex import iex
import os
```

```python
# Set parameters
period = 'quarter'
time = '12'
symbol = ["FSLY"]
token = os.environ['TOKEN']
```

```python
# Get and preprocess data
portfolio = iex.get_df(symbol, period, time, token)
portfolio_det = iex.add_company(symbol, token)
portfolio = iex.preprocess(portfolio)
portfolio = iex.add_profitability(portfolio)
portfolio = iex.add_financial_strength(portfolio)
```

```python
# Financial graphs portfolio
iex.graphs_financials(portfolio)
```

```python
# Profitability graphs portfolio
iex.graphs_profitability(portfolio)
```

```python
# Financial security graphs portfolio
iex.graphs_security(portfolio)
```
